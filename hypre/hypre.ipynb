```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
refills = pd.read_csv('hypre-offline-profiling.txt')
```

```python
fig, ax = plt.subplots(1, dpi=300)
plt.yscale('log')
ax.plot(refills['Index'], refills['cpu0_refill'], label="CPU 0")
ax.plot(refills['Index'], refills['cpu1_refill'], label="CPU 1")
ax.plot(refills['Index'], refills['cpu2_refill'], label="CPU 2")
ax.legend(loc='best')
ax.set_title("Offline Profiling for Hypre ex13, -n 3500, 3 ranks")
ax.set_xlabel("Time (seconds)")
ax.set_ylabel("Cache Refills (unit)")
```
