```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid", palette="pastel")

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('overhead.csv')
df = df
# Create a separate boxplot for each application with its own y-scale
g = sns.FacetGrid(df, col='app', col_wrap=3, height=4, sharey=False)
g.map(sns.boxplot, 'type', 'dif')
g.fig.suptitle("Overhead Comparison between Kub and Volcano-native Runs")
# Add labels and title to each plot
for ax, app in zip(g.axes.flat, df['app'].unique()):
    ax.set_xlabel('Type')
    ax.set_ylabel('Execution Time (s)')
    ax.set_title(f'Application: {app}')

# Adjust spacing between plots
g.tight_layout()

# Display the plots
plt.savefig("overhead.jpg", dpi=300)
```

```python
cm1 = pd.read_csv('cm1.csv')
gmx = pd.read_csv('gmx.csv')
stream = pd.read_csv('smpi.csv')
```

```python

# Read the data from CSV file
data = pd.read_csv('alla.csv')

# Create a barplot
sns.set(style='whitegrid')
plt.subplots(3, 1, figsize=(10, 12))
plt.suptitle("Behaviour of Different Applications for Elastic Scaling Patterns")

# Plot two barplots side by side for each 'app'
plt.subplot(3, 1, 1)
sns.barplot(data=cm1, x='scen', y='time', ci='sd', hue='case', palette='pastel', hue_order=['C1', 'C2', 'C3'])
plt.title('Application: cm1')
plt.xlabel('scenario')
plt.ylabel('execution time (s)')

plt.subplot(3, 1, 2)
sns.barplot(data=gmx, x='scen', y='time', ci='sd', hue='case', palette='pastel', hue_order=['C1', 'C2', 'C3'])
plt.title('Application: gromacs')
plt.xlabel('scenario')
plt.ylabel('execution time (s)')

plt.subplot(3, 1, 3)
sns.barplot(data=stream, x='scen', y='time', ci='sd', hue='case', palette='pastel', hue_order=['C1', 'C2', 'C3'])
plt.title('Application: stream-mpi')
plt.xlabel('scenario')
plt.ylabel('execution time (s)')

# Adjust the layout and display the plot
plt.tight_layout()
plt.savefig("apps.jpg", dpi=300)
plt.show()
```
